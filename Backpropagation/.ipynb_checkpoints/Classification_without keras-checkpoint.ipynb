```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,12,0]], columns=['cgpa', 'score', 'lpa'])
```

```python
def initialize_parameters(layer_dims):
    np.random.seed(3)
    parameters={}
    L=len(layer_dims)

    for l in range (1, L):
        parameters['w' + str(l)] = np.ones((layer_dims[l-1],layer_dims[l]))*0.1
        parameters['b' + str(l)] = np.zeros((layer_dims[l],1))
    return parameters
        
```

```python
def sigmoid(z):
    A=1/(1+np.exp(-z))
    # A=sigmoid(z)
    return A
```

```python
# initialize_parameters([2,2,1])
```

```python
def linear_forward(A_prev , w, b):
    z=np.dot(w.T, A_prev) +b
    A=sigmoid(z)
    
    return A
```

```python
# forward propagation
def L_layer_forward(x, parameters):
    A=x
    L=len(parameters)//2
    for l in range (1,L+1):
        A_prev= A
        w1= parameters['w'+ str(1)]
        b1= parameters['b'+ str(1)]
        # print('A'+str(l-1)+':', A_prev)
        # print('w'+str(l)+':', w1)
        # print('b'+str(l)+':', b1)
        # print('--'*20)

        A=linear_forward(A_prev , w1 , b1)
        # print('A'+str(l)+':', A)
        # print('**'*20)
    return A, A_prev  # it gives me y_hat ,o11 and o12
        
        
```

```python
def update_parameter(parameters,y,y_hat,A1,x):
    parameters['w2'][0][0]=parameters['w2'][0][0]+(0.001 * (y-y_hat)*A1[0][0])
    parameters['w2'][1][0]=parameters['w2'][1][0]+(0.001*(y-y_hat)*A1[1][0])
    parameters['b2'][0][0]=parameters['w2'][1][0]+(0.001*(y-y_hat))

    parameters['w1'][0][0]=parameters['w1'][0][0]+(0.001 *(y-y_hat)*parameters['w2'][0][0]*A1[0][0]*(1-A1[0][0])*x[0][0])
    parameters['w1'][0][1]=parameters['w1'][0][1]+(0.001 *(y-y_hat)*parameters['w2'][0][0]*A1[0][0]*(1-A1[0][0])*x[1][0])
    parameters['b1'][0][0]=parameters['b1'][0][0]+(0.001 *(y-y_hat)*parameters['w2'][0][0]*A1[0][0]*(1-A1[0][0]))

    parameters['w1'][1][0]=parameters['w1'][1][0]+(0.001 *(y-y_hat)*parameters['w2'][1][0]*A1[1][0]*(1-A1[1][0])*x[0][0])
    parameters['w1'][1][1]=parameters['w1'][1][1]+(0.001 *(y-y_hat)*parameters['w2'][1][0]*A1[1][0]*(1-A1[1][0])*x[1][0])
    parameters['b1'][1][0]=parameters['b1'][1][0]+(0.001 *(y-y_hat)*parameters['w2'][1][0]*A1[1][0]*(1-A1[1][0]))

                                                   
```

```python
x=df[['cgpa', 'score']].values[0].reshape(2,1)
y=df[['lpa']].values[0][0]

parameters=initialize_parameters([2,2,1])
# parameters
y_hat,A1=L_layer_forward(x, parameters)
y_hat=y_hat[0][0]
update_parameter(parameters, y , y_hat , A1, x)
print('loss for this std' , -y*np.log(y_hat)-(1-y)*np.log(1-y_hat))
parameters
```

```python
# epocha implementation 
parameters= initialize_parameters([2,2,1])
epochs=100

for i in range(epochs):

    loss=[]
    for j in range(df.shape[0]):
        x=df[['cgpa', 'score']].values[j].reshape(2,1)
        y=df[['lpa']].values[j][0]

        y_hat, A1 =L_layer_forward(x, parameters)
        y_hat= y_hat[0][0]

        update_parameter(parameters, y, y_hat, A1 , x)

        loss.append(-y*np.log(y_hat)-(1-y)*np.log(1-y_hat))
    print('Epoch - ', i+1, 'Loss - ' , np.array(loss).mean())
parameters
```
