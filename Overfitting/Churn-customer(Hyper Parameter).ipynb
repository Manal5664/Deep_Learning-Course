```python
x=df.iloc[:,:-1]
```

```python
y=df['Exited']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit_transform(x)
```

```python
x=pd.DataFrame(ss.fit_transform(x) , columns=x.columns )
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x , y , test_size=0.2, random_state=10)
```

```python
x_test
```

## Overfitting Early Stopping and Batch Normalization

```python
import tensorflow
from keras.layers import Dense, BatchNormalization , Dropout
from keras.regularizers import L2
from keras.models import Sequential
from keras.callbacks import EarlyStopping
```

```python
ann=Sequential()
```

```python
ann.add(Dense(6 , input_dim=8 , activation='relu', kernel_regularizer=L2(l2=0.01)))
ann.add(Dense(5, activation='relu', kernel_regularizer=L2(l2=0.01)))
ann.add(Dense(3, activation='relu', kernel_regularizer=L2(l2=0.01)))
ann.add(Dense(3, activation='relu', kernel_regularizer=L2(l2=0.01)))
ann.add(Dense(2 , activation='relu', kernel_regularizer=L2(l2=0.01)))
ann.add(Dense(1 ,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam' , loss='binary_crossentropy' , metrics=['accuracy'])
```

```python
ann.fit(x_train , y_train , batch_size=10 , epochs=50 , validation_data=(x_test , y_test), callbacks=EarlyStopping())     
```

```python
train_acc=ann.history.history['accuracy']
# training accuracy after validation we got testing too
test_acc=ann.history.history['val_accuracy']
```

```python
len(test_acc)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([i for i in range(1,35)], train_acc)
plt.plot([i for i in range(1,35)], test_acc, c='red')
plt.show()
```

```python
pred=ann.predict(x_test)
```

```python
prd_data= []
for i in pred:
    if i[0]>0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```
